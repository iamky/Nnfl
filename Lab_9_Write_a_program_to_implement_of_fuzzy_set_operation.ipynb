```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

x = np.arange(0, 11, 1)

A = fuzz.trimf(x, [0, 5, 10])
B = fuzz.trimf(x, [0, 3, 6])
```

```python
union = np.fmax(A, B)
intersection = np.fmin(A, B)
complement_A = 1 - A

print("Union:", union)
print("Intersection:", intersection)
print("Complement of A:", complement_A)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, A, label='A(x)', linewidth=2)
plt.plot(x, B, label='B(x)', linewidth=2)
plt.plot(x, union, label='A ∪ B', linestyle='--')
plt.plot(x, intersection, label='A ∩ B', linestyle='--')
plt.plot(x, complement_A, label="A'", linestyle=':')
plt.title("Fuzzy Set Operations")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid(True)
plt.show()
```
