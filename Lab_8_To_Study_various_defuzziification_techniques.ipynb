```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
# Universe of discourse
x = np.arange(0, 11, 1)

# Triangular membership function: trimf(x, [a, b, c])
mf = fuzz.trimf(x, [0, 5, 10])

plt.figure(figsize=(6,4))
plt.plot(x, mf, linewidth=2)
plt.title("Membership Function for Defuzzification")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.grid(True)
plt.show()
```

```python
centroid = fuzz.defuzz(x, mf, 'centroid')   # Center of Gravity
bisector = fuzz.defuzz(x, mf, 'bisector')   # Vertical bisector of area
mom      = fuzz.defuzz(x, mf, 'mom')        # Mean of Maxima
som      = fuzz.defuzz(x, mf, 'som')        # Smallest Maxima
lom      = fuzz.defuzz(x, mf, 'lom')        # Largest Maxima

print("Centroid:", centroid)
print("Bisector:", bisector)
print("MOM:", mom)
print("SOM:", som)
print("LOM:", lom)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, mf, label='Membership Function', linewidth=2)

# Plot all crisp defuzzified values
plt.axvline(centroid, color='red', linestyle='--', label=f'Centroid = {centroid:.2f}')
plt.axvline(bisector, color='green', linestyle='--', label=f'Bisector = {bisector:.2f}')
plt.axvline(mom, color='blue', linestyle=':', label=f'MOM = {mom:.2f}')
plt.axvline(som, color='purple', linestyle=':', label=f'SOM = {som:.2f}')
plt.axvline(lom, color='orange', linestyle=':', label=f'LOM = {lom:.2f}')

plt.title("Comparison of Defuzzification Techniques")
plt.xlabel("x")
plt.ylabel("Membership Degree")
plt.legend()
plt.grid(True)
plt.show()
```
